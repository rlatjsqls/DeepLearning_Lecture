```python
# 붓꽃 데이터 세트의 feature들 확인
print('\n feature_names 의 type:', type(iris_data.feature_names))
print(' feature_names 의 shape:', len(iris_data.feature_names))
print(iris_data.feature_names)

# 붓꽃 데이터 세트의 target이 무엇인지 확인
print('\n target_names 의 type:',type(iris_data.target_names))
print(' feature_names 의 shape:',len(iris_data.target_names))
print(iris_data.target_names)

# 전체 data의 데이터타입과 구조, 저장된 값 확인
print('\n data 의 type:',type(iris_data.data))
print(' data 의 shape:',iris_data.data.shape)
print(iris_data['data'])

# target 데이터의 데이터 타입, 형태, 저장된 값 확인
print('\n target 의 type:',type(iris_data.target))
print(' target 의 shape:',iris_data.target.shape)
print(iris_data.target)

# feature = 우리가 학습시킬 자료.
# dataframe에서 하나의 column

# target = 우리가 목표로 하는 정답지
# feature를 가지고 학습시켜서 target을 예측/분류 하는것을 목표로 함
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

# 붓꽃 데이터 세트를 로딩합니다.
iris = load_iris()

# iris.data 변수에 저장.
# iris.data는 Iris 데이터 세트에서 피처(feature)만으로 된 데이터를 numpy로 가지고 있습니다.
iris_data = iris.data

# iris.target 변수에 저장.
# iris.target은 붓꽃 데이터 세트에서 레이블(결정 값) 데이터를 numpy로 가지고 있습니다.
iris_label = iris.target
print('iris target값:', iris_label)

# iris target명 확인 0='setosa', 1='versicolor', 2='virginica'
print("iris target명:",iris.target_names)

# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환합니다.
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head()
```

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)    #무작위로 조건을 mapping 랜덤시드를 고정할 수 있음

# 모델 학습 시작
dt_clf.fit(iris_data, iris_label)

# 학습 데이터 셋으로 예측 수행
pred = dt_clf.predict(iris_data)

# 예측 정확도를 구하기 위한 accuracy_score import
from sklearn.metrics import accuracy_score
                                    #정답지     예상답안
print('예측 정확도:',accuracy_score(iris_label, pred)) # 가지고있는 정답 데이터 대비, 예측한 답안이 얼마나 맞았는지 비율
```

## train_test_split

```python
# data, label을 train, test 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)

# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 수행
dt_clf.fit(X_train, y_train)

# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행.
pred = dt_clf.predict(X_test)

# 에측해보기
from sklearn.metrics import accuracy_score
print('예측 정확도: {0:4f}'.format(accuracy_score(y_test,pred)))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/월수업/머신러닝 실습용자료/InkjetDB_preprocessing.csv")

X = df.drop(['PatternSize'], axis = 1)
y = df['PatternSize']
df
```

## 학습 및 테스트 데이터세트 분리

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)

print('학습 데이터 :',X_train.shape)
print('테스트 데이터 :', X_test.shape)
print('학습 데이터 :',y_train.shape)
print('테스트 데이터 :', y_test.shape)
```

## 선형회귀(Linear Regression)

- 결정계수
- 평균제곱근오차

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

#모델정의
model = LinearRegression()

#모델학습
model.fit(X_train, y_train)

#예측
y_pred = model.predict(X_test)

#결정 계수 계산
r2 = r2_score(y_test, y_pred)

#평균 제곱근 오차 계산
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("결정 계수:",r2)
print("평균 제곱근 오차:",rmse)
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error

KNN_model = KNeighborsRegressor()

#학습
KNN_model.fit(X_train, y_train)

#예측
y_pred = KNN_model.predict(X_test)

#결정 계수 계산
KNN_r2 = r2_score(y_test, y_pred)

#평균 제곱근 오차 계산
KNN_mse = mean_squared_error(y_test, y_pred)
KNN_rmse = np.sqrt(KNN_mse)

print("결정 계수:",KNN_r2)
print("평균 제곱근 오차:",KNN_rmse)
```

## SVM

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

SVM_model = SVR()

#학습
SVM_model.fit(X_train, y_train)

#예측
y_pred = SVM_model.predict(X_test)

#결정 계수 계산
SVM_r2 = r2_score(y_test, y_pred)

#평균 제곱근 오차 계산
SVM_mse = mean_squared_error(y_test, y_pred)
SVM_rmse = np.sqrt(SVM_mse)

print("결정 계수:",SVM_r2)
print("평균 제곱근 오차:",SVM_rmse)
```

```python
# data, target 정의
data = study['공부시간']
target = study['시험점수']

# 산점도 그리기
import matplotlib.pyplot as plt
plt.plot(data,target,'o')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# train, test 데이터 나누기 위해 numpy로 변경
data = study['공부시간'].to_numpy()
target = study['시험점수'].to_numpy()

# 훈련 세트와 테스트 세트로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=40)
```

```python
# shape 확인
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# reshape 함수에 -1을 넣으면, rows은 자동으로 입력해달라는 뜻
# 즉, column을 1개로 내가 정했으니, 남은 데이터는 알아서 row에 잘 넣어주세요.
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
```

```python
# shape 확인
X_train
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# 점수 확인(lr.score는 결정계수R2에 대한 정확도를 의미합니다.)
print(model.score(X_train , y_train))
print(model.score(X_test , y_test))
```

```python
# 16이라는 값을 넣었을 때 예상 결과값 확인
model.predict([[16]])
```

```python
# 회귀계수 확인
#y = Wx + b (W값, b값)
print(model.coef_ , model.intercept_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train , y_train)
plt.plot( [5,18], [5*model.coef_ +model.intercept_ ,
                    18*model.coef_ + model.intercept_])
plt.scatter(16 , 90 ,marker="^")
plt.show()
```

```python
# 다항회귀분석 적용
import numpy as np
#column을 늘려주는 함수
훈련용_data_poly = np.column_stack(( X_train ** 2, y_train))
테스트용_data_poly = np.column_stack((X_test ** 2 , y_test))

lr = LinearRegression()
lr.fit(훈련용_data_poly , y_train)
lr.score(테스트용_data_poly , y_test)
```

```python
lr.predict([[16**2,16]])
```
